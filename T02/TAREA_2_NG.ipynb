Hacer un programa que impirma los numeros del -100 al 200 que sean divisibles entre 6 (con resto 0)

```python
# EJERCICIO 02

numero_inicial = -100
numero_final = 201
lista = range(numero_inicial, numero_final)
for i in lista:
  if i % 6 == 0:
    print(i)
```
